```python
import plot.route as pr
pr.plot_vehicle_routes(instance, solution, jointly=True, figsize=(10,10), show_arrows=False, show_node_labels=False)
```

```python
import matplotlib.pyplot as plt
n_plots = len(instance.vehicles)
fig, ax = plt.subplots(n_plots, figsize=(10,100))

v_nodes=dict()
for v in solution.vehicle_routes:
    v_nodes[v.id] = v.visits

cmap = pr.get_cmap(len(v_nodes))

for i, (v, visits) in enumerate(v_nodes.items()):
    ax[i].set_title(instance.vehicle_id_dict[v].alias)
    pr.plot_vehicle_route(ax[i], visits, instance.node_id_dict, route_color=cmap(i))
```
